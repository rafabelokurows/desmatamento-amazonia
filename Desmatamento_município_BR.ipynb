# Campeões do desmatamento

Objetivo: desenvolver um web app ou dashboard simples com alguns indicadores sobre desmatamento e infrações ambientais no Brasil

## Fontes de dados
* INPE (Prodes)
* Ibama

## Indicadores

### Desmatamento
* Municípios de maior área (e área de Amazônia)
* Mapa BR com os top N municípios
* Municípios onde ocorreu maior perda de área de floresta (Amazônica)

### Infrações 
* TOP N municípios em infrações Ibama
* TOP N indivíduos/empresas
* Infrações por ano/mês
* Infrações Amazônia Legal (por município)

```python
import pandas as pd
```

```python
desmatamento = pd.read_csv('/content/gdrive/MyDrive/desmatamento.csv')
municipios = pd.read_csv('/content/gdrive/MyDrive/municipios.csv')
```

```python
desmatamento.head()
municipios.info()
```

```python
df=pd.merge(desmatamento,municipios,how='left').loc[:,["ano","id_municipio","nome","area","desmatado","incremento","floresta","nuvem","nao_observado","hidrografia","nao_floresta","nome_uf"]]
df['pctdesmatado'] = df.desmatado / df.floresta
df.head()
#df['incrementopct'] = df.pctdesmatado / 
```

```python
df.query("ano == 2021").sort_values('area',ascending=False).head(10)
```

```python
df.query("ano == 2021").sort_values('desmatado',ascending=False).head(10)
```

```python
df.query("ano == 2021").sort_values('pctdesmatado',ascending=False).head(50)
```

```python
df['diffdesmatado'] = df.sort_values(['nome','nome_uf','ano']).groupby(['nome','nome_uf'])['desmatado'].diff(1)
df.groupby('ano').diffdesmatado.sum()
# df.query("ano == 2021").sort_values('diffdesmatado',ascending=False).head()
# df.query("nome == 'Candeias do Jamari'")
```
